```python
t = 0
for area in data["Area"].unique():
    d = data[data["Area"] == area]
    d = d[d["Item"] == "Population"]
    try:
        plt.title("population of people in " + area +  " over the years")
        plt.bar(d.columns[4:], d.iloc[0,4:])
        plt.show()
        t += 1
    except Exception as e:
        print("Error : ", e)
        print(area + " has an issue")
        
print("plotted graphs for " + str(t) + " areas")
    
```
